## IQR-Based Outlier Handling

```python
# Select columns for outlier analysis
column = df[['age','campaign','duration']]

# Calculate IQR bounds
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Filter values within bounds
df[['age','campaign','duration']] = column[(column > lower_bound) & (column < upper_bound)]
```

```python
# Create Box-plot
df.plot(kind='box', subplots=True, layout=(2,5),figsize=(20,10),color='#808000')
plt.suptitle('Feature Distributions After Outlier Treatment', y=1.01, fontsize=16)
plt.show()
```

# Label Encoding Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df_encoded = df1.apply(lb.fit_transform)
df_encoded
```

# Data Preparation for Machine Learning

## Target Variable Analysis

```python
# Check class distribution in target variable
df_encoded['deposit'].value_counts()
```

```python
# Split into independent (features) and dependent (target) variables
x = df_encoded.drop('deposit',axis=1)  # independent variable
y = df_encoded['deposit']              # dependent variable

# Verify shapes and types
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

# Train-Test Split for Model Development

## Data Partitioning Configuration

```python
from sklearn.model_selection import train_test_split
```

```python
# Split data into training (75%) and testing (25%) sets
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)

# Print dimensions of resulting datasets
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Model Evaluation Utilities

```python

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

# Performance Evaluation Function
def eval_model(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

# Model Scoring Function
def mscore(model):
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    print('Training Score',train_score)  
    print('Testing Score',test_score)
```

# Decision Tree Classifier Implementation

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize Decision Tree with specified parameters
dt = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)

# Train the model on training data
dt.fit(x_train,y_train)
```

```python
# Calculate training and test scores
mscore(dt)
```

```python
# Generate predictions on test set
ypred_dt = dt.predict(x_test)
print(ypred_dt)
```

```python
# Evaluate model performance
eval_model(y_test,ypred_dt)
```

```python
from sklearn.tree import plot_tree
```
